# Series To Dataframe

Dataframe is created by using the pandas `Series.to_frame()` function and `pandas.DataFrame()` function.

## How to create a dataframe from pandas series. 

Series is a 1-d labeled array and can hold any data type.

Dataframe is a collection of series.

In this article, you will learn how to create a dataFrame from series using -

* built-in series function.
* pandas dataframe function.


Let's see the syntax for creating a pandas series.

### pandas.Series()
 
__pandas.Series(data=None, index=None, dtype=None, name=None, copy=False)__

__Purpose:__ This function is used for creating a pandas series.


__Parameters:__

 __- data:__ ndarray, iterable, dict.

 __- index:__ index to use for resulting frame.

 __- dtype:__ datatype.

 __- name:__ series name.

 __- copy:__ copy data from inputs.


__Returns:__ Pandas series.

## 1.Creating dataframe from built-in series function `Series.to_frame()`


````python
# Importing library
import pandas as pd

fruits = ['Apple','Banana','Guava','Blueberries','Grapes']

# Creating Series
s1 = pd.Series(fruits, name="Fruit")     
df = s1.to_frame()

print("Type:",type(df))
df
````

##2.Creating dataframe from single or multiple series using `pandas.DataFrame()`


````python
color = ['Red','Yellow','Green','Blue','Violet']

# Creating a series for above list
s2 = pd.Series(color, name="colors")

# Creating a dataframe by passing series to dataframe's values
df = pd.DataFrame({'Fruit': s1, 'Color': s2})

print("Type:",type(df))
df
````

## 3.Practical Tips 

* ```Series.to_frame()```  method  is limited to a particular series. So use ```pandas.DataFrame()``` method.
