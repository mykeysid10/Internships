## Procedure: 

> Original Image → Gray Scale → Invert → Blur → Invert → Pencil Sketch



## Importing Libraries



```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

## Reading The Input Image 

```python
img = cv2.imread("/content/drive/MyDrive/LGMVIP/input_image.png")
img = cv2.resize(img,(240,240))
cv2_imshow(img)
```

## Converting Original Image to Grey Scale

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

## Inverting the Gray Image

```python
#inv_gray_img = 255 - gray_img
inv_gray_img = cv2.bitwise_not(gray_img)
cv2_imshow(inv_gray_img)
```

## Blurring the inverted gray image by Gausian Function

```python
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
cv2_imshow(blur_img)
```

## Inverting the blurred image

```python
#inv_blur_img = 255 - blur_img
inv_blur_img = cv2.bitwise_not(blur_img)
cv2_imshow(inv_blur_img)
```

## Creating Pencil Sketch image

```python
pencil_img = cv2.divide(gray_img,inv_blur_img,scale = 256.0)
cv2_imshow(pencil_img)
```

```python
cv2.imwrite("/content/drive/MyDrive/LGMVIP/output_image.png",pencil_img)
```

## Conclusion 

> Converted a RGB image to pencil sketch using Python.
